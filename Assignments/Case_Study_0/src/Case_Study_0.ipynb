# Step7: Identify outliers
* In general outliers can be detected by using the 4 meithods
    * boxlpot
    * qqplot
    * IQR (Most widely used)
    * z-score

1. Outliers are of 2 types
    * LO
    * UO
* L.O. (lower outliers)= Q1 - 3 * IQR (this is practically)
* U.O. (upper outliers)= Q3 + 3 * IQR (this is practically)
* iqr = Q3-Q1
* Mild outliers: for research mild outliers (we wont consider in real time). If huge diff b/w 50% to 75%, 25% to 50%
* Wild outliers(LO, UO) mu - 3 sigma to mu + 3 sigma 
* B/c of Mild outliers we have data NOT normally distributed hence we apply variable transformation technique
    * square
    * sqrt
    * Log
    * Inverse
